#### Example 1: Boston Housing Data Set 

```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
#import csv library
bos_hou_df = pd.read_csv('BostonHousing_02_01_2023.csv')
bos_hou_df
```

```python
#define dataframe for new name
bh_df = bos_hou_df
bh_df
```

```python
#CAT.MEDV here represents the categorized median
#rename CAT.MEDV for easier data handling
bh_df = bh_df.rename(columns={'CAT. MEDV':'CAT_MEDV'})
bh_df.head(9)
```

```python
#statistics implementation
bh_df.describe()
```

```python
#see the described data separated
print('Mean : ', bh_df.CRIM.mean())
print('Std. dev.: ', bh_df.CRIM.std())
print('Min : ', bh_df.CRIM.min())
print('Max : ', bh_df.CRIM.max())
print('Median : ', bh_df.CRIM.median())
print('Length : ', len(bh_df.CRIM))
print('Number of missing values: ', bh_df.CRIM.isnull().sum())
```

```python
pd.DataFrame({'mean' : bh_df.mean(),
'Sd' : bh_df.std(),
'min' : bh_df.min(),
'max' : bh_df.max(),
'median' : bh_df.median(),
'length' : len(bh_df),
'miss.val': bh_df.isnull().sum(),
})
```

```python
bh_df.corr().round(2)
```

```python
bh_df['RM_bin'] = pd.cut(bh_df.RM, range(0, 10), labels=False)
print(bh_df['RM_bin'].head(9), bh_df['RM'].head(9))
```

```python
bh_df.groupby(['RM_bin', 'CHAS'])['MEDV'].mean()
```

```python
pd.pivot_table(bh_df, values='MEDV', index=['RM_bin'],
columns=['CHAS'], aggfunc=np.mean, margins=True)
```

```python
#using method crosstab to crate a cross-tabulation of two variables
tb1 = pd.crosstab(bh_df.CAT_MEDV, bh_df.ZN)
print(tb1)
```

```python
#convert number to ratios 
propTb1 = tb1 / tb1.sum()
propTb1.round(2)
```

```python
#plot the ratios in a stacked bar chart
ax = propTb1.transpose().plot(kind='bar', stacked=True)
plt.title('Distribution of CAT_MEDV by ZN')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
